```python
#dataset-raw.csv is taken from kaggle, 
# with the title of the dataset "cause-of-death-in-indonesia"
import pandas as pd
df = pd.read_csv('dataset-raw.csv') 
dfDeath = df[["Total Deaths"]].sum()
print("sum of people who died\n",dfDeath) #sum
dfYearDeath = df[["Year","Total Deaths"]].groupby("Year").mean()
print("Average death per year\n",dfYearDeath) #mean
dfMostDeath = df.groupby("Cause").sum().mode()
print("Most Death\n", dfMostDeath) #mode
dfMedianYear = df[["Year"]].median()
print("Median\n",dfMedianYear) #median
dfVariant = df[["Cause","Total Deaths"]].groupby("Cause").sum("Total Deaths").var()
print("Variant\n",dfVariant) #variant
dfSD = df[["Cause","Total Deaths"]].groupby("Cause").sum("Total Deaths").std()
print("STD\n",dfSD) #standard deviation
```

```python
import matplotlib.pyplot as plt
df = pd.DataFrame(df)
newdf = df[(df.Year==2015)]
X = list(newdf.iloc[:, 0])
plt.xticks(rotation=90)
Y = list(newdf.iloc[:, 4])
plt.bar(X, Y, color='g')
plt.title("Number of death with its cause in 2015")
plt.xlabel("cause of death")
plt.ylabel("total death")
```
